```python
import pandas as pd
import ast


MAX_TIMESTAMP = 1e12


# ==============================
# 1. CODE CŨ
# ==============================

def transform_age_old(
    df: pd.DataFrame,
    date: str
) -> pd.DataFrame:

    clean_df = df[["cus_id", "ngay_sinh"]].copy()

    clean_df = clean_df.dropna(subset=["cus_id"])


    # xử lý cus_id giống code mới
    clean_df["cus_id"] = clean_df["cus_id"].apply(
        lambda x: ast.literal_eval(x).get("member0")
        if isinstance(x, str) and x.startswith("{")
        else (x.get("member0") if isinstance(x, dict) else x)
    )


    clean_df["cus_id"] = (
        clean_df["cus_id"]
        .astype(float)
        .astype(int)
        .astype(str)
    )


    # tính tuổi
    dob = pd.to_numeric(
        clean_df["ngay_sinh"],
        errors="coerce"
    )


    dob = dob.apply(
        lambda x: x if x < MAX_TIMESTAMP else None
    )


    dob = pd.to_datetime(
        dob,
        unit="ms",
        errors="coerce"
    )


    ref_date = pd.to_datetime(date)


    clean_df["nam_tuoi"] = (
        (ref_date - dob)
        .dt.days / 365.25
    )


    # giống logic cũ nếu có duplicate cus_id
    old_df = (
        clean_df
        .groupby("cus_id")
        .agg({
            "nam_tuoi": "max"
        })
        .reset_index()
    )


    return old_df[
        [
            "cus_id",
            "nam_tuoi"
        ]
    ]


# ==============================
# 2. CODE MỚI
# copy đúng logic của bạn
# ==============================

def transform_age_new(
    df: pd.DataFrame,
    date: str
) -> pd.DataFrame:


    clean_df = df[["cus_id", "ngay_sinh"]].copy()

    clean_df = clean_df.dropna(subset=["cus_id"])


    clean_df["cus_id"] = clean_df["cus_id"].apply(
        lambda x: ast.literal_eval(x).get("member0")
        if isinstance(x, str) and x.startswith("{")
        else (x.get("member0") if isinstance(x, dict) else x)
    )
    clean_df["cus_id"] = (
        clean_df["cus_id"]
        .astype(float)
        .astype(int)
        .astype(str)
    )
    dob = pd.to_numeric(
        clean_df["ngay_sinh"],
        errors="coerce"
    )
    dob = dob.apply(
        lambda x: x if x < MAX_TIMESTAMP else None
    )
    dob = pd.to_datetime(
        dob,
        unit="ms",
        errors="coerce"
    )
    ref_date = pd.to_datetime(date)

    clean_df["age_years"] = (
        (ref_date - dob)
        .dt.days / 365.25
    )
    new_df = clean_df[
        [
            "cus_id",
            "age_years"
        ]
    ]

    return new_df

# ==============================
# 3. LOAD DATA RAW
# ==============================
raw_df = raw_df.copy()

# ngày tính tuổi
date = "2025-12-31"

# ==============================
# 4. CHẠY 2 VERSION
# ==============================

old_result = transform_age_old(
    raw_df,
    date
)

new_result = transform_age_new(
    raw_df,
    date
)

# ==============================
# 5. ĐỔI TÊN ĐỂ SO SÁNH
# ==============================

new_result = new_result.rename(
    columns={
        "age_years": "nam_tuoi"
    }
)
# ==============================
# 6. COMPARE
# ==============================

compare = old_result.merge(
    new_result,
    on="cus_id",
    how="outer",
    suffixes=("_old", "_new")
)

# check mất cus
print(
    "Cus chỉ có old:",
    compare["nam_tuoi_new"].isna().sum()
)


print(
    "Cus chỉ có new:",
    compare["nam_tuoi_old"].isna().sum()
)

# check giá trị tuổi khác
diff = compare[
    (
        compare["nam_tuoi_old"].round(6)!= compare["nam_tuoi_new"].round(6)
    )
    &
    ~(
        compare["nam_tuoi_old"].isna()&compare["nam_tuoi_new"].isna()
    )
]
print(
    "Số cus có tuổi khác:",
    len(diff)
)
print(diff.head(20))

compare_check = compare.copy()

compare_check["same"] = (
    compare_check["nam_tuoi_old"].fillna(-999)==compare_check["nam_tuoi_new"].fillna(-999)
)
print(
    "Số cus giống nhau:",
    compare_check["same"].sum()
)
print(
    "Số cus khác nhau:",
    (~compare_check["same"]).sum()
)
```

```python
import pandas as pd
import numpy as np
# ============================================================
# 1. COMMON COMPARE FUNCTION
# ============================================================

def compare_feature(
    old_df,
    new_df,
    feature_name,
    feature_cols
):

    print("\n")
    print("="*100)
    print(feature_name)
    print("="*100)

    # --------------------------
    # Schema
    # --------------------------

    print("\nSCHEMA")

    print("OLD:")
    print(old_df.columns.tolist())

    print("NEW:")
    print(new_df.columns.tolist())

    # --------------------------
    # datatype
    # --------------------------

    print("\nDATATYPE")

    print("OLD")
    print(old_df.dtypes)

    print("\nNEW")
    print(new_df.dtypes)

    # --------------------------
    # duplicate
    # --------------------------

    print("\nDUPLICATE")

    print(
        "old duplicate:",
        old_df["cus_id"].duplicated().sum()
    )

    print(
        "new duplicate:",
        new_df["cus_id"].duplicated().sum()
    )
    # --------------------------
    # customer count
    # --------------------------

    print("\nCUSTOMER")

    print(
        "old cus:",
        old_df["cus_id"].nunique()
    )

    print(
        "new cus:",
        new_df["cus_id"].nunique()
    )

    # --------------------------
    # cus difference
    # --------------------------

    old_cus = set(
        old_df["cus_id"].astype(str)
    )

    new_cus = set(
        new_df["cus_id"].astype(str)
    )

    print("\nCUS DIFFERENCE")

    print(
        "Only old:",
        len(old_cus-new_cus)
    )

    print(
        "Only new:",
        len(new_cus-old_cus)
    )

    # --------------------------
    # merge compare
    # --------------------------

    compare = old_df.merge(
        new_df,
        on="cus_id",
        how="outer",
        suffixes=("_old","_new")
    )

    print("\nVALUE DIFFERENCE")


    for col in feature_cols:
        old_col = col+"_old"
        new_col = col+"_new"

        if old_col not in compare.columns:
            print(
                col,
                "missing old"
            )
            continue

        if new_col not in compare.columns:
            print(
                col,
                "missing new"
            )
            continue

        same = (
            compare[old_col]
            .fillna(-999999)
            ==
            compare[new_col]
            .fillna(-999999)
        )

        diff = (~same).sum()

        print(
            col,
            "different:",
            diff
        )

        if diff > 0:

            print(
                compare.loc[
                    ~same,
                    [
                        "cus_id",
                        old_col,
                        new_col
                    ]
                ]
                .head(10)
            )

    # --------------------------
    # null
    # --------------------------

    print("\nNULL")

    print("OLD")

    print(
        old_df[feature_cols]
        .isna()
        .sum()
    )

    print("\nNEW")

    print(
        new_df[feature_cols]
        .isna()
        .sum()
    )

    # --------------------------
    # statistic
    # --------------------------

    print("\nSTATISTICS OLD")

    print(
        old_df[feature_cols]
        .describe()
    )

    print("\nSTATISTICS NEW")

    print(
        new_df[feature_cols]
        .describe()
    )


    return compare
```

```python

# ============================================================
# 2. SALES REGION
# ============================================================
valid_region = [
    # list 48 tỉnh của bạn
]

def old_sales_region(
    df,
    column_name
):

    temp = df.copy()

    temp["month"] = pd.to_datetime(
        temp["month"],
        format="%Y%m"
    ).dt.to_period("M")

    latest = (
        temp
        .groupby("cus_id")["month"]
        .max()
        .reset_index()
    )

    temp = temp.merge(
        latest,
        on="cus_id"
    )

    result = (
        temp[
            temp["month"]
            ==
            temp["month_y"]
        ]
        [
            [
                "cus_id",
                column_name
            ]
        ]
        .drop_duplicates()
    )

    result["sales_region"] = (
        result[column_name]
        .isin(valid_region)
        .astype(int)
    )

    return (
        result
        .groupby("cus_id")
        ["sales_region"]
        .max()
        .reset_index()
    )

def new_sales_region(
    df,
    column_name
):

    temp = df.copy()


    temp["month"] = pd.to_datetime(
        temp["month"],
        format="%Y%m"
    ).dt.to_period("M")

    latest_month = (
        temp.groupby("cus_id")
        ["month"]
        .transform("max")
    )

    result = (
        temp[
            temp["month"]
            ==
            latest_month
        ]
        [
            [
                "cus_id",
                column_name
            ]
        ]
        .drop_duplicates()
    )

    result["sales_region"] = (
        result[column_name]
        .isin(valid_region)
        .astype(int)
    )

    return (
        result
        .groupby("cus_id")
        ["sales_region"]
        .max()
        .reset_index()
    )

# chạy
path = "clean/fts/sales_region/month/partition=202512/data.parquet"

clean_month_df = read_minio_parquet(path)

clean_month_df.head()
clean_month_df.columns.tolist()
old = old_sales_region(
    clean_month_df,
    "ma_tinh_hdong_chinh"
)


new = new_sales_region(
    clean_month_df,
    "ma_tinh_hdong_chinh"
)


compare_feature(
    old,
    new,
    "SALES REGION",
    [
        "sales_region"
    ]
)

# ============================================================
# 3. USAGE DURATION
# ============================================================


def old_usage_duration(df):

    result = (
        df.groupby("cus_id")
        ["usage_months"]
        .max()
        .reset_index()
    )

    return result.rename(
        columns={
            "usage_months":
            "usage_months_feature"
        }
    )



def new_usage_duration(df):

    return (
        df[
            [
                "cus_id",
                "usage_months"
            ]
        ]
        .rename(
            columns={
                "usage_months":
                "usage_months_feature"
            }
        )
    )



compare_feature(
    old_usage_duration(clean_month_df),
    new_usage_duration(clean_month_df),
    "USAGE DURATION",
    [
        "usage_months_feature"
    ]
)





# ============================================================
# 4. ORDER AVG VALUE
# ============================================================


def old_avg_value(df):

    temp = df.copy()

    temp["month"] = pd.to_datetime(
        temp["month"],
        format="%Y%m"
    ).dt.to_period("M")


    max_month=temp["month"].max()


    temp=temp[
        temp["month"]
        .between(
            max_month-5,
            max_month
        )
    ]


    result = (
        temp.groupby("cus_id")
        .agg(
            total_value=("tong_tien","sum"),
            months=("month","nunique")
        )
        .reset_index()
    )


    result["avg_value"] = (
        result["total_value"]
        /
        result["months"]
    )


    return result[
        [
            "cus_id",
            "avg_value"
        ]
    ]



def new_avg_value(df):

    temp=df.copy()

    temp["month"] = pd.to_datetime(
        temp["month"],
        format="%Y%m"
    ).dt.to_period("M")


    max_month=temp["month"].max()


    temp=temp[
        temp["month"]
        .between(
            max_month-5,
            max_month
        )
    ]


    result=(
        temp
        .groupby(
            [
                "cus_id",
                "month"
            ]
        )
        .agg(
            monthly_value=("tong_tien","sum")
        )
        .groupby("cus_id")
        .agg(
            avg_value=("monthly_value","mean")
        )
        .reset_index()
    )


    return result



compare_feature(
    old_avg_value(clean_month_df),
    new_avg_value(clean_month_df),
    "AVG SUCCESS ORDER VALUE",
    [
        "avg_value"
    ]
)
```
