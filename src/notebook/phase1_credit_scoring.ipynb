# II. THIẾT KẾ FEATURE TABLE

## 1. Xử lý `ngay_sinh` (dob)

**Thực trạng dữ liệu**
`ngay_sinh`:
- Missing rất nhiều
- Kiểu `float64`
- Có giá trị dạng epoch milliseconds. Ví dụ: 455155200000

Không thể bỏ cột này vì:
- Dùng để suy ra age merchant
- Age là biến quan trọng trong scoring

**Quy tắc xử lý**

| Trường hợp         | Cách xử lý              |
| ------------------ | ----------------------- |
| `ngay_sinh` = null | Fill = **avg tuổi**     |
| `ngay_sinh` hợp lệ | Chuẩn hoá → date → tuổi |

````python
def normalize_dob(df):
    df = df.copy()

    # 1. ép về numeric để loại string / object / dict
    dob_raw = pd.to_numeric(df['ngay_sinh'], errors='coerce')

    # 2. loại bỏ inf / -inf (BẮT BUỘC)
    dob_raw = dob_raw.replace([np.inf, -np.inf], np.nan)

    # 3. giới hạn khoảng DOB hợp lý (1900-01-01 → 2010-01-01)
    min_ms = pd.Timestamp("1900-01-01").value // 10**6
    max_ms = pd.Timestamp("2010-01-01").value // 10**6

    # ⚠️ CLIP thay vì where → không overflow
    dob_raw = dob_raw.clip(lower=min_ms, upper=max_ms)

    # 4. convert sang datetime (AN TOÀN)
    df['dob'] = pd.to_datetime(
        dob_raw,
        unit='ms',
        errors='coerce'
    )

    # 5. tính tuổi
    today = pd.Timestamp.today()
    df['age_years'] = ((today - df['dob']).dt.days / 365.25)

    # 6. fill missing bằng tuổi trung bình
    avg_age = df['age_years'].mean()
    df['age_years'] = df['age_years'].fillna(avg_age)

    return df

day = normalize_dob(day)
day.head()
````

## 2. Chuẩn hoá khoá định danh merchant

Trong dữ liệu gốc:
- cus_id là một object dạng dict
- ID thực tế nằm ở cus_id['member0']

→ Ta chuẩn hoá thành merchant_id để dùng xuyên suốt.

````python
def extract_merchant_id(x):
    if isinstance(x, dict):
        return x.get("member0")
    return None

day["merchant_id"] = day["cus_id"].apply(extract_merchant_id)
month["merchant_id"] = month["cus_id"].apply(extract_merchant_id)
````

## 3. Join dữ liệu day & month

**Logic**
- Base table: daily snapshot
- Monthly data: bổ sung hành vi dài hạn
- Join: LEFT JOIN

````python
df = day.merge(
    month[
        [
            "merchant_id",
            "tongdon_cao",
            "tongdon_thap",
            "khachhang_quaylai",
            "khachhang_roibo"
        ]
    ],
    on="merchant_id",
    how="left"
)

df.head()
````

**Quy ước Phase 1**

````python
month_cols = [
    "tongdon_cao",
    "tongdon_thap",
    "khachhang_quaylai",
    "khachhang_roibo"
]

df[month_cols] = df[month_cols].fillna(0)
````

## 5. Thời gian hợp tác (Usage Duration)

Ý nghĩa
- Merchant hợp tác lâu → ổn định hơn
- Feature quan trọng cho credit scoring

````python
df["ngay_hoptac_dt"] = pd.to_datetime(
    df["ngay_hoptac"],
    dayfirst=True,
    errors="coerce"
)

df["usage_duration_months"] = (
    (datetime.today() - df["ngay_hoptac_dt"]).dt.days / 30
)

df["usage_duration_months"].describe()
````

# III. RULE-BASED ELIGIBILITY (THEO CHUẨN VTP)

Áp dụng đúng policy VTP để xác định:
- Merchant ĐỦ ĐIỀU KIỆN hay KHÔNG
- Nếu không đạt → fail ở tiêu chí nào

**Đặc điểm**
- Hard rule (Boolean)
- Không cộng điểm
- Không bù trừ
- Fail 1 điều kiện = fail toàn bộ

## 1. Tổng hợp rule VTP (official)

| STT | Tiêu chí                           | Điều kiện                |
| --- | ---------------------------------- | ------------------------ |
| 1   | Tuổi                               | 20 ≤ age ≤ 50            |
| 2   | Thời gian sử dụng                  | ≥ 6 tháng                |
| 3   | Địa bàn                            | ≥ 48 tỉnh                |
| 4   | Doanh thu TB/tháng (6 tháng)       | ≥ 10 triệu               |
| 5   | Không có tháng < 3 triệu (6 tháng) | min ≥ 3 triệu            |
| 6   | Số đơn TB/tháng (6 tháng)          | ≥ 10                     |
| 7   | Doanh thu 3 quý                    | Không giảm >30% liên tục |
| 8   | Doanh thu 4 tháng                  | Không giảm >20% liên tục |

**LƯU Ý RẤT QUAN TRỌNG VỀ DATA**

Dataset hiện tại:
- **KHÔNG** có time series theo tháng
- **KHÔNG** có province count
- **KHÔNG** có quarterly revenue

Vì vậy:
- Các rule **CÓ** data → triển khai trực tiếp
- Các rule **CHƯA** có data → đánh dấu NOT_EVALUATED

→ Đây là cách làm đúng trong thực tế, không được tự suy diễn. Sẽ bổ sung trong phase 2.

## 2. Khởi tạo bảng rule check

````python
rule_df = df.copy()
````

## 3. RULE 1 – Tuổi (20–50)

````python
rule_df["rule_age"] = (
    (rule_df["age_years"] >= 20)
    & (rule_df["age_years"] <= 50)
)
````

## 4. RULE 2 – Thời gian sử dụng ≥ 6 tháng

````python
rule_df["rule_usage_duration"] = (
    rule_df["usage_duration_months"] >= 6
)
````

## 5. RULE 3 – Địa bàn thuộc 48 tỉnh (CHƯA THỂ ĐỐI CHIẾU)

Hiện tại:
- Chỉ có `ma_tinh_hoatdong_chinh`

→ Rule này sẽ được triển khai khi có bảng valid_provinces.

## 6. RULE 4 – Doanh thu TB ≥ 10tr/tháng (6 tháng)

Yêu cầu:
```bash
    sum(revenue_last_6m) / 6 ≥ 10,000,000
```

Hiện tại:
- `tong_tien` = lifetime
- Không có window 6 tháng

````python
rule_df["rule_avg_monthly_revenue_6m"] = None
````

## 7. RULE 5 – Không có tháng < 3tr (6 tháng)

Yêu cầu:
```bash
    min(monthly_revenue_last_6m) ≥ 3,000,000
```

````python
rule_df["rule_min_monthly_revenue_6m"] = None
````

## 8. RULE 6 – Số đơn TB/tháng ≥ 10 (6 tháng)

````python
rule_df["rule_avg_monthly_orders_6m"] = None
````

## 10. RULE 8 – Doanh thu 4 tháng không giảm >20%

````python
rule_df["rule_monthly_decline_4m"] = None
````

## 11. Tổng hợp kết quả rule

### a. Xác định rule CÓ THỂ ĐÁNH GIÁ

````python
evaluated_rules = [
    "rule_age",
    "rule_usage_duration"
]
````

### b. Merchant đạt tất cả rule đã đánh giá?

````python
rule_df["eligible_partial"] = rule_df[evaluated_rules].all(axis=1)
rule_df.head()
````

## 12. Ghi rõ lý do FAIL

````python
def fail_reasons(row):
    reasons = []

    if not row["rule_age"]:
        reasons.append("Tuổi không trong khoảng 20–50")

    if not row["rule_usage_duration"]:
        reasons.append("Thời gian sử dụng < 6 tháng")

    return "; ".join(reasons)

rule_df["fail_reason"] = rule_df.apply(fail_reasons, axis=1)
````

````python
rule_df[
    [
        "merchant_id",
        "eligible_partial",
        "fail_reason"
    ]
].head(10)
````
