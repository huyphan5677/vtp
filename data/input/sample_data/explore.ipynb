```python
# Làm sạch cus_id: chuyển từ dict string sang giá trị member0
import ast

def extract_cus_id(cus_id_val):
    if isinstance(cus_id_val, dict):
        return cus_id_val.get('member0')
    elif isinstance(cus_id_val, str):
        try:
            d = ast.literal_eval(cus_id_val)
            return d.get('member0')
        except:
            return cus_id_val
    return cus_id_val

df["cus_id_clean"] = df["cus_id"].apply(extract_cus_id)
df_test = df.groupby(["cus_id_clean"]).agg({"ma_tinh_hoatdong_chinh": "nunique"}).reset_index()
```
